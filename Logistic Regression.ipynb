```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
h_data = pd.read_csv('housing.csv')
h_data = h_data.dropna()
h_data = h_data.drop(h_data.loc[h_data['median_house_value'] == 500001].index)
h_data = pd.get_dummies(h_data, columns=['ocean_proximity'])
```

```python
h_data.sample(5)
```

```python
median = h_data['median_house_value'].median()
```

```python
h_data['above_median'] = (h_data['median_house_value'] - median) > 0
```

```python
h_data.sample(5)
```

```python
X = h_data.drop(['median_house_value', 'above_median'], axis=1)
Y = h_data['above_median']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(solver='liblinear').fit(x_train, y_train)
print("Training Score : ", logistic_model.score(x_train, y_train))
```

```python
y_pred = logistic_model.predict(x_test)
df_pred_actual = pd.DataFrame({'predicted': y_pred, 'actual': y_test})
df_pred_actual.head(10)
```
