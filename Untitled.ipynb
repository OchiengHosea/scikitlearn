```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
h_data = pd.read_csv('housing.csv')
h_data = h_data.dropna()
```

```python
h_data.shape
```

```python
h_data = h_data.drop(h_data.loc[h_data['median_house_value'] == 500001].index) # remove outlier / skewed values
```

```python
h_data.shape
```

```python
h_data = pd.get_dummies(h_data, columns=['ocean_proximity']) # Convert categorical vales to numeric
```

```python
h_data.sample(5)
```

```python
X = h_data.drop('median_house_value', axis=1)
Y = h_data['median_house_value']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression(normalize=True).fit(x_train, y_train) # normalizing tranforms data to be between 0 to 1
```

```python
print("Training_score ", linear_model.score(x_train, y_train))
```

```python
predictors = x_train.columns
coef = pd.Series(linear_model.coef_, predictors).sort_values()
print(coef) #this shows the level of influence of each variable in the prediction
```

```python
y_pred = linear_model.predict(x_test)
```

```python
# Model performance evaluation
df_pred_actual = pd.DataFrame({'predicted': y_pred, 'actual': y_test})
df_pred_actual.head(10)
```

```python
# Evaluate using R2
from sklearn.metrics import r2_score
print("Testing Score : ", r2_score(y_test, y_pred))
```

```python
# use scatterplot to visualize actual vs predicted values
fig, ax = plt.subplots(figsize=(12, 8))
plt.scatter(y_test, y_pred)
plt.show()
```

```python
# Visualize using a line chart
df_pred_actual_sample = df_pred_actual.sample(100)
df_pred_actual_sample = df_pred_actual_sample.reset_index()
df_pred_actual_sample.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_pred_actual_sample['predicted'], label='Predicted')
plt.plot(df_pred_actual_sample['actual'], label='Actual')
plt.ylabel('median_house_value')
plt.legend()
plt.show()
```
